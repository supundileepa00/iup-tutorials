### Plot Image

```python
Colorimg = cv.imread('lena.png', cv.IMREAD_COLOR)
Grayimg = cv.imread('lena.png', cv.IMREAD_GRAYSCALE)
Unchangedimg = cv.imread('lena.png', cv.IMREAD_UNCHANGED)
fig, ax = plt.subplots(1,3,figsize = (15,15))
ax[0].imshow(cv.cvtColor(Colorimg,cv.COLOR_BGR2RGB))
ax[0].axis('on')
ax[1].imshow(cv.cvtColor(Grayimg,cv.COLOR_BGR2RGB))
ax[1].axis('on')
ax[2].imshow(cv.cvtColor(Unchangedimg,cv.COLOR_BGR2RGB))
ax[2].axis('on')
```

```python
plt.imshow(cv.cvtColor(Unchangedimg,cv.COLOR_BGR2RGB))
plt.axis('on')
```
